```python
import pandas as pd
```

```python
with open(r"C:\Users\aamankos\Downloads\Telegram Desktop\ACNEDet v1.v2i.yolokeras\train\_annotations.txt", "r") as file:
    fl = file.read()
arr = fl.split("\n")
```

```python
arr_2d = []
for i in arr:
    arr_2d.append(i.split(" "))
arr_2d
```

```python
overall=[]
q=0
for i in range(len(arr_2d)):
    arr_3d=[]
    arr_3d.append(arr_2d[i][0])
    for j in range(len(arr_2d[i])):
        if j != 0:
            temp = arr_2d[i][j].split(",")[-1]
            arr_3d.append(temp)
    overall.append(arr_3d)
overall
```

```python
data = {
    'Name': [],
    '0': [],
    '1': [],
    '2': [],
    '3': []
}

df = pd.DataFrame(data)

for i in overall:
    temp_data={}
    temp_data['Name'] = i[0]
    if '0' in i:
        temp_data['0'] = 1
    else:
        temp_data['0'] = 0
    if '1' in i:
        temp_data['1'] = 1
    else:
        temp_data['1'] = 0
    if '2' in i:
        temp_data['2'] = 1
    else:
        temp_data['2'] = 0
    if '3' in i:
        temp_data['3'] = 1
    else:
        temp_data['3'] = 0
    df.loc[len(df)] = temp_data
df
```

```python
df.to_csv("classes_seperated_train.csv", index = False)
```
